# now let's train linear reg model on data

````python
x=df['RM']
````

````python
x
````

### split the data

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=.2)
````

````python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
````

````python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
````

````python
y_train=np.reshape(y_train,(-1,1))
print(y_train.shape)
````

````python
x_train = np.array(x_train).reshape( -1,1)
````

````python
Lr.fit(x_train,y_train)
````

````python
x_test=np.array(x_test).reshape(-1,1)
````

````python
y_hat=Lr.predict(x_test)
````

````python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
````

## visulazition traing set

````python
Lr.predict(x_train).shape
````

````python
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,Lr.predict(x_train))
plt.xlabel('Av num of rooms')
plt.ylabel('price')
plt.title('num of rooms vs price')
plt.show()
````

## visulazition test set

````python
plt.scatter(x_test,y_test,color='blue')
plt.plot(x_train,Lr.predict(x_train))
plt.xlabel('Av num of rooms')
plt.ylabel('price')
plt.title('num of rooms vs price')
plt.show()
````

## score and error of model

````python
Lr.score(x_train,y_train)
````

````python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('Mean absolute error : ',mean_absolute_error(y_hat,y_test))

print('Mean squared error : ',mean_squared_error(y_hat,y_test))
````
